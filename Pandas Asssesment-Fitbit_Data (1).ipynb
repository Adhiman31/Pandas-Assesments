```python
#9. how many person are not a regular person with respect to activity try to find out those 
```

```python
fb
```

```python
fb['TotalSteps']
```

```python
act = fb[["ActivityDate","Id"]].groupby("Id").count().reset_index()
def act_status(a):
    for i in act['ActivityDate']:
        if a < 10:
            return 'Irregular'
        elif a >10 & a<25:
            return 'Regular'
        elif a >= 25 :
            return 'Very Active'
act['ActivityStatus']= act['ActivityDate'].apply(act_status)
act = act[act['ActivityStatus']=='Irregular']['Id']
act
len(act)
```
